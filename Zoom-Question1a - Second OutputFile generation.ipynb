```python
import pandas as pd
import json
from datetime import datetime
```

```python
data = pd.read_json('meeting.txt')
```

```python
data.head()
```

```python
meetingid_total_persons_frame = data[['meetingid', 'userid']].drop_duplicates()['meetingid'].value_counts().rename_axis('meetingid').reset_index(name='total_persons')
```

```python
meetingid_total_persons_frame
```

```python
host_data = data[data.is_host==True]
```

```python
host_data
```

```python
meetingid_hosts_frame = host_data[['meetingid', 'userid']].drop_duplicates()['meetingid'].value_counts().rename_axis('meetingid').reset_index(name='no_hosts')
```

```python
attendee_data = data[data.is_host==False]
```

```python
attendee_data
```

```python
meetingid_attendee_frame = attendee_data[['meetingid', 'userid']].drop_duplicates()['meetingid'].value_counts().rename_axis('meetingid').reset_index(name='no_attendees')
```

```python
date_format = '%Y-%m-%d %H:%M:%S'
only_date_format = '%Y-%m-%d'
str_2_date_fn = lambda x: datetime.strptime(x, date_format)
data['jointime'] = data['jointime'].apply(str_2_date_fn)
data['leavetime'] = data['leavetime'].apply(str_2_date_fn)
data['date'] = data['leavetime'].apply(lambda x: x.strftime(only_date_format))
data['duration'] = data['leavetime'] - data['jointime']
data['duration_secs'] = data['duration'].dt.total_seconds()
data['duration_mins'] = data['duration'].dt.total_seconds().div(60)
```

```python
total_seconds_frame = data.groupby('meetingid').duration_secs.sum().rename_axis('meetingid').reset_index(name='total_seconds')
```

```python
total_seconds_frame
```

```python
total_mins_frame = data.groupby('meetingid').duration_mins.sum().rename_axis('meetingid').reset_index(name='total_minutes')
```

```python
total_mins_frame
```

```python
data.head()
```

```python
meetingid_date_country_company_frame = data[['meetingid', 'date', 'country', 'Company']].drop_duplicates()
```

```python
meetingid_date_country_company_frame
```

```python
host_attendee_frame = pd.merge(meetingid_hosts_frame, meetingid_attendee_frame)
```

```python
host_attendee_frame
```

```python
minutes_secs_frame = pd.merge(total_mins_frame, total_seconds_frame)
```

```python
minutes_secs_frame
```

```python
host_time_frame = pd.merge(host_attendee_frame, minutes_secs_frame)
```

```python
host_time_frame
```

```python
res = pd.merge(host_time_frame, meetingid_date_country_company_frame)
```

```python
res
```

```python
output  = res.to_json(orient='records')
import json
with open('output1_test3.txt', 'w') as outfile:
    json.dump(output, outfile, indent=4)
```
